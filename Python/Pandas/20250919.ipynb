```python
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
Custom_df= pd.DataFrame(pd.read_csv('/kaggle/input/bigdatacertificationkr/Customer_Data.csv'))
```

```python
# 고개 데이터의 컬럼들 확인
Custom_df.columns


# 데이터 프레임의 각 컬럼들의 타입과 널 값의 존재 유무
Custom_df.info()

# DataFrame에서 내가 원하는 컬럼을 기준으로하여 내림차순 정렬하기
# sort_values 함수를 사용하며 파라미털로는 by= "내가 원하는 컬럼 명", ascending = True(올림차순) / False(내림차순)
income_sorted = Custom_df.sort_values(by='income', ascending= False)

# 수입이 5번쨰로 많은 사람의 수입을 가져오기
fifth_largest_income = income_sorted.iloc[4]['income']
```

```python
# 아이가 3명 이상인 사람들의 평균 수입이 얼마인가
# 소수점 2자리수까지 표현해라

Custom_df[Custom_df['children']>=3]['income'].mean().round(decimals=2)
```

```python
# 구매한 사람들의 평균 수입이 얼마인가
# 소수점 2자리수까지 표현해라

Custom_df[Custom_df['purchase'] == 1]['income'].mean().round(decimals=2)
```

```python
# 구매한 사람들의 평균 나이가 얼마인가
# 소수점 2자리수까지 표현해라

Custom_df[Custom_df['purchase'] == 1]['age'].mean().round(decimals=2)
```

```python
# 구매한 사람들 중 성별이 남자인 사람은 몇명인가? (남자는 gender=1 이라고 한다.)
# 'age'라는 컬럼의 널값이 없기 때문에 age의 수를 센다.

Custom_df[(Custom_df['purchase'] == 1) & (Custom_df['gender']==1)]['age'].count()
```
